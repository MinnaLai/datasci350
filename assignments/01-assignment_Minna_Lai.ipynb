### Question 01

Convert the decimal number 53 to binary. Show your work.

2^5 = 32, which represents 100000 <br> 
53-32 = 21 which contains 2^4 = 16 (10000)<br> 
21 - 16 = 4 + 1, which represents 100 + 1 <br> 
53 = 32 + 16 + 4 + 1 = 2^5 + 2^4 + 2^2 + 2^0 <br> 
So 53 in binary is 100000+10000+101=110101

### Question 08

Describe the Von Neumann architecture and its significance in modern computing. What is the Von Neumann bottleneck, and how does it affect computer performance?

- The Von Neumann architecture is the computer architecture that stores program instructions and data in a slow storage medium like hard disk and transfers them to faster RAM for CPU to execute.
- It is significant because it is the basis for almost all modern computers, and it is efficient in memory use, flexible in data storage and manipulation, and simple in design and operation.
- The Von Neumann bottleneck is the limits of computing performance due to the single bus processing of instructions and data sequentially.
- It lowers the computing performance due to the faster processing speed of the CPU compared to the memory transfer rate, so CPU has to wait for data.

### Question 09

Compare and contrast low-level and high-level programming languages. Give two examples of each and explain when you might choose to use one over the other.

- Both low-level and high-level programming languages allow humans to write instructions that will be executed by a computer.
- Low-level programming languages are closer to machine code and harder for human to read and write, but they provide more direct control over the hardware; High-level languages are closer to human languages and abstract from hardware details that are portable across different systems.
- Low level language examples are Machine Code and Assembly Language, which are used when precise hardware or operating system control is required.
- High level language examples are Python and Java, which are used for more general software development or data analysis.

### Question 10

Discuss the concept of abstraction in computer science, using the representation of images in computers as an example. How does this abstraction process impact data analysis and predictive modelling in image-related tasks?

Abstraction in computer science is the process of simplifying the complicated (often real-world) things into storable and analyzable digital number, so that computers can store and process. In order to represent images in computers, we abstract the image into pixel (each dot in a grid of colors), which can be stored numerically corresponding to the color. This abstraction enables the predictive modelling to analyze the patterns in pixel values arrangement, so that the images can be classified or predicted based on the abstracted data of pictures rather than the complicated raw images in real life.
